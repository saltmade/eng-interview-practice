# Valid Sudoku

Given a 9x9 grid `board` representing a Sudoku, return `true` if the board does not have any conflicts and `false` otherwise. The board contains only numbers between `0` and `9` in each cell, with `0`s denoting empty cells. A *conflict* is a duplicate number along a row, a column, or a 3x3 subgrid. For the purposes of this problem, it doesn't matter if the Sudoku has a valid solution or not â€” only whether it has a conflict in the already-filled cells.

Example
```
1 0 0 0 0 0 1 0 0

```

Algo 1
- Walk the grid, finding values and evaluating each
    - search rows
    - search cols
    - search in subgrid
- Divvy up subgrids
    - 0-2, 3-5, 6-8 ^2
    - (0-2, 0-2), ()
    - 2, 2 

Algo 2
- Walk all rows, columns and subgrids
    - Use a set to keep track of seen values
n = length of one side of the sudoku
T: O(1) (or O(n^2), but the input is always a fixed size so it defeats the purpose of describing complexity)
S: O(1) (same)

````python
def is_valid(board):
    def valid_rows():
        R, C = len(board), len(board[0])
        for r in range(R):
            seen = set()
            for c in range(C):
                val = board[r][c]
                if val in seen:
                    return False
                elif val:
                    seen.add(val)

        return True

    def valid_columns():
        R, C = len(board), len(board[0])
        for c in range(C):
            seen = set()
            for r in range(R):
                val = board[r][c]
                if val in seen:
                    return False
                elif val:
                    seen.add(val)

        return True

    def valid_subgrids():
        def valid_subgrid(R, C):
            seen = set()
            for r in range (R, R + 3):
                for c in range(C, C + 3):
                    val = board[r][c]
                    if val in seen:
                        return False
                    elif val:
                        seen.add(val)

        for r in [0, 3, 6]:
            for c in [0, 3, 6]:
                if not valid_subgrid(r, c):
                    return False
        
        return True

    return valid_rows() and valid_columns() and valid_subgrids()
````
